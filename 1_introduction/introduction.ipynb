### クラス
これまでintやstrなどのデータ型を見てきた(`type()`という関数でオブジェクトの型を調べられる)．それらのデータ型は "組み込み" のデータ型といい，初めからpythonに組み込まれたデータ型．

ここでは，新しいクラスを定義する．ユーザーが新しいクラスを自分で定義すれば，独自にデータ型を作成することができる．

また，オリジナルのメソッド（クラス用の関数）や属性を定義することも可能．

pythonでは`class`というキーワードを使って，クラスを定義する．
クラスのフォーマット
```
class クラス名:
    def __init__(self, 引数, ...):  # コンストラクタ
        ...
    def メソッド名１(self, 引数, ...):  # メソッド1
        ...
    def メソッド名２(self, 引数, ...):  # メソッド2
        ...
```
`__init__`は，初期化を行う特別なメソッド．"コンストラクタ(constructor)"とも呼ばれ，クラスのインスタンスが作成される際に1度だけ呼ばれる．
また，pythonではメソッドの第一引数に自分自身（自身のインスタンス）を表す self を明示的に書くのが特徴．ほかの言語からするとこれは奇妙らしい．．．

#### Numpyのインポート

Numpyは外部ライブラリ．

#### Numpy配列の生成
 `np.array()`というメソッドを使ってNumPyの配列を生成する．
 `np.array()`はpythonのリストを引数に取り，Numpy用の配列 (numpy.ndarray)を作成する．

 #### Numpyの算術計算
以下の計算で注意すべき点は，配列のxとyの要素数が同じであるという点．(両者とも要素数が3の1次元配列)
xとyの要素数が同じ場合，算術計算は各要素に対して行われる．異なる場合はエラー
"要素ごと"は英語で"element-wise"という
Numpy配列とスカラ値の組み合わせでも算術計算可能．

#### ブロードキャスト
Numpyでは形状の異なる配列の演算も可能．

### Matplotlib
グラフを描画するためには，matplotlibのpyplotというモジュールを使う．

````python
# Matplotlib
import numpy as np
import matplotlib.pyplot as plt
````

````python
# sin関数の描画
# データの作成
x = np.arange(0, 6, 0.1) # 0から6まで0.1刻みで生成
y = np.sin(x)
````

````python
# グラフの描画
plt.plot(x, y) # x, yのデータ列をplt.plotメソッドに与える
plt.show() # グラフを表示
````

````python
# pyplot の機能
x = np.arange(0, 6, 0.1) # 0から6まで0.1刻みで生成
y1 = np.sin(x)
y2 = np.cos(x)

# グラフの描画
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label ="cos") # 破線で描写
plt.xlabel("x") # x軸のラベル
plt.ylabel("y") # y軸のラベル
plt.title('sin & cos') # タイトル
plt.legend()
plt.show()
````
